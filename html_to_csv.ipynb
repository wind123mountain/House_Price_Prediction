### **Key - Value encode categori**

```python
import json

with open('D:\\MachineLearning\\DataScience\\House Price Prediction\\raw_data.json', 'r') as f:
    data = json.load(f)
```

```python
ft = [ 'street_id', 'property_road_condition', 'property_back_condition', 'ward', 
       'furnishing_sell', 'house_type', 'property_legal_document', 'area', 'direction', 'area_v2', 'street_id']

features = {}
for key in ft:
    features[key] = {}
```

```python
for indt in data.values():
    for key in ft:
        try:
            if key == 'area_v2':
                features[key][indt['ad']['area_v2']] = indt['ad_params']['area']['value']
                continue

            val = indt['ad'][key]
            if key == 'street_id':
                features[key][val] = indt['ad']['address']
            else:
                features[key][val] = indt['ad_params'][key]['value']
            
        except:()
    
```

```python
with open('D:\\MachineLearning\\DataScience\\House Price Prediction\\House_Price_Prediction\\raw_data\\encode_features_value.json', 'w', encoding='utf-8') as f:
    json.dump(features, f)
```

### **Covert html to json, csv**

```python
import os

# src = 'data_4'
src = ''
html_lists = os.listdir(src)
html_lists[:5]
```

```python
del_list = ['images', 'videos', 'ad_id', 'date', 'account_id', 'account_oid', 'account_name',
            'category_name', 'area_name', 'region_name', 'company_ad', 'price_string', 'type',
            'number_of_images', 'avatar', 'ward_name', 'pty_map', 'streetnumber_display',
            'contain_videos', 'phone', 'type_name', 'reviewer_image', 'reviewer_nickname',
            'thumbnail_image', 'address', 'owner','protection_entitlement', 'unitnumber_display',
            'has_video']
```

```python
from bs4 import BeautifulSoup
import json
import codecs
import time

item = {}
col_list = set()

for file in html_lists:
    f=codecs.open(src + '/' + file, 'r', 'utf-8')
    soup = BeautifulSoup(f.read(), "html.parser")
    f.close()

    out = soup.find(id="__NEXT_DATA__")
    data = json.loads(out.string)
    indt = data['props']['initialState']['adView']['adInfo']
    del indt['headers']
    del indt['parameters']
    del indt['ad_params']
    for key in del_list:
        try:
            del indt['ad'][key]
        except:()
    
    id = str(indt['ad'].pop('list_id'))
    item[id] = indt['ad']
    item[id]['list_time'] = time.strftime("%d %B %Y", time.localtime(item[id]['list_time']/1000))

    col_list.update(item[id])
    
```

```python
len(item)
```

```python
len(col_list)
```

```python
import json

with open('raw_data/raw_data_4.json', 'w', encoding='utf-8') as f:
    json.dump(item, f)
```

```python
import numpy as np

pre_data = {}

for col in col_list:
    pre_data[col] = {}

import numpy as np

for key in item:
    for col in col_list:
        try:
            pre_data[col][key] = item[key][col]
        except:
            pre_data[col][key] = np.nan
```

```python
import pandas as pd

df = pd.DataFrame(pre_data)

df
```

```python
df.to_csv('pre_data/pre_data_4.csv', index_label = 'id')
```

### **Load raw_data**

```python
import json

with open('D:\\MachineLearning\\DataScience\\House Price Prediction\\raw_data.json', 'r') as f:
    raw_data = json.load(f)
```

```python
del_list = ['images', 'videos', 'ad_id', 'date', 'account_id', 'account_oid', 'account_name',
            'category_name', 'region_name', 'company_ad', 'type', 'unitnumber', 'has_video',
            'number_of_images', 'avatar', 'pty_map', 'streetnumber_display', 'category',
            'contain_videos', 'phone', 'type_name', 'reviewer_image', 'reviewer_nickname',
            'thumbnail_image', 'owner','protection_entitlement', 'unitnumber_display']
```

```python
import time

item = []
col_list = set()

for indt in raw_data:
    del raw_data[indt]['parameters']
    del raw_data[indt]['ad_params']

    for key in del_list:
        try:
            del raw_data[indt]['ad'][key]
        except:()
    
    raw_data[indt]['ad']['list_time'] = time.strftime("%d %B %Y", time.localtime(raw_data[indt]['ad']['list_time']/1000))

    item.append(raw_data[indt]['ad'])

    col_list.update(raw_data[indt]['ad'])
```

```python
import pandas as pd

df = pd.DataFrame(item)
```
